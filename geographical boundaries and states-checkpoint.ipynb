```python
#plotting both boundries twogather 
import matplotlib.pyplot as plt
ax = indian_boundry.plot(color='none', edgecolor='red', linewidth=1.5, figsize=(12, 12))
india_states.plot(ax=ax, color='lightgreen', edgecolor='black', linewidth=0.5)
plt.title("India Boundary and States Overlay")
plt.show()
```

```python
print("Boundary CRS:", indian_boundry.crs)
print("States CRS:", india_states.crs)

if indian_boundry.crs != indian_boundry.crs:
    indian_boundry = indian_boundry.to_crs(india_states.crs)
    print("CRS aligned.")
```

```python
ax = indian_boundry.plot(color='none', edgecolor='red', linewidth=1.5, figsize=(12, 12))
india_states.plot(ax=ax, color='lightgreen', edgecolor='black', linewidth=0.5)
plt.title("State Boundaries Within National Boundary")
plt.show()
```

```python
print(india_states.columns)
```

```python
india_states['area_sqkm'] = india_states.geometry.area / 10**6  # Convert from square meters
print(india_states[['st_nm', 'area_sqkm']])
```

```python
india_states['boundary_length_km'] = india_states.geometry.length / 1000  # Convert from meters
print(india_states[['st_nm', 'boundary_length_km']])
```

```python
states_near_boundary = india_states[india_states.geometry.intersects(indian_boundry.unary_union)]

# Display these states
print(states_near_boundary[['st_nm']])
```

```python
# Perform a spatial join to combine boundary and state data
joined_gdf = india_states.sjoin(indian_boundry, how="inner", op="intersects")

joined_gdf.head(4)
```

```python
# Plot the India States with national boundary
fig, ax = plt.subplots(figsize=(12, 12))

# Plot the India States with some transparency
india_states.plot(ax=ax, color='lightblue', edgecolor='black', alpha=0.5)

# Plot the National Boundary
indian_boundry.plot(ax=ax, color='red', edgecolor='black', alpha=0.7, linewidth=2)

# Add state names to the map
for idx, row in india_states.iterrows():
    ax.annotate(row['st_nm'], xy=row['geometry'].centroid.coords[0], 
                xytext=(3, 3), textcoords='offset points', fontsize=8)

plt.title("Indian States and National Boundary with State Names")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```

```python
# Plot the India States with national boundary
fig, ax = plt.subplots(figsize=(12, 12))

# Plot the India States with some transparency
india_states.plot(ax=ax, color='lightblue', edgecolor='black', alpha=0.5)

# Plot the National Boundary
indian_boundry.plot(ax=ax, color='red', edgecolor='black', alpha=0.7, linewidth=2)

# Add state names, area, and boundary length to the map
for idx, row in india_states.iterrows():
    state_name = row['st_nm']
    area = round(row['area_sqkm'], 2)  # Area in square km
    boundary_length = round(row['boundary_length_km'], 2)  # Boundary length in km
    
    # Annotate with state name, area, and boundary length
    label = f"{state_name}\nArea: {area} sq km\nBoundary Length: {boundary_length} km"
    ax.annotate(label, xy=row['geometry'].centroid.coords[0], 
                xytext=(3, 3), textcoords='offset points', fontsize=8)

plt.title("Indian States and National Boundary with Area and Boundary Length")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```

```python
# Color coding by area size (for example, states with larger areas will be darker)
fig, ax = plt.subplots(figsize=(12, 12))

# Use area as a criterion for color coding
india_states.plot(ax=ax, column='area_sqkm', cmap='YlGnBu', legend=True, 
                  legend_kwds={'label': "State Area (sq km)", 'orientation': "horizontal"})

# Plot the National Boundary
indian_boundry.plot(ax=ax, color='grey', edgecolor='black', alpha=0.7, linewidth=2)

plt.title("Indian States with Area Color Coding and National Boundary")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Prepare the data for clustering
data = india_states[['area_sqkm', 'boundary_length_km']]
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Perform K-means clustering
kmeans = KMeans(n_clusters=5)
india_states['cluster'] = kmeans.fit_predict(scaled_data)

# Visualize the clusters
fig, ax = plt.subplots(figsize=(12, 12))
india_states.plot(ax=ax, column='cluster', cmap='viridis', legend=True, 
                  legend_kwds={'label': "State Clusters", 'orientation': "horizontal"})
indian_boundry.plot(ax=ax, color='pink', edgecolor='black', alpha=0.7, linewidth=2)
plt.title("K-means Clustering of Indian States")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```
